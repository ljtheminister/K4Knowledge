```python
import pandas as pd
import os
from collections import defaultdict
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas.tools.rplot as rplot
import cPickle as pickle
from ggplot import *
```

```python
base_dir = os.getcwd() + '/'
data = pd.read_csv(base_dir + 'astudentData.csv')
grouped = data.groupby('question_id')
```

```python
Q_counts = defaultdict(int)
Q_correct = defaultdict(int)
Q = data['question_id'].unique()

for row_idx in xrange(len(data)):
    x = data.ix[row_idx,:]
    id = x['question_id']
    Q_counts[id] += 1
    Q_correct[id] += x['correct']

Q_correct_pct = dict()
for q in Q:
     Q_correct_pct[q] = Q_correct[q]/float(Q_counts[q])

'''
pickle.dump(Q_correct_pct, open('Q_correct_pct.pkl', 'wb'))      
Q_correct_pct = pd.read_pickle('Q_correct_pct.pkl')
'''
```

```python
q_var = []
for q_id in grouped.groups.keys():
    q_var.append(np.var(data.ix[ grouped.groups[q_id] , 'correct' ]))
    
q_var = pd.Series(data=q_var, index=grouped.groups.keys())
q_var.sort(ascending=False) #find questions with highest variance
```

```python
q_accuracy = []
q_variance = [] #re-fix variance to be in in-class variances, right now they are total variances
for q_id in q_var.index:
    q_accuracy.append(np.mean(data.ix[ grouped.groups[q_id] , 'correct' ]))
    q_variance.append(np.var(data.ix[ grouped.groups[q_id] , 'correct' ]))
plt.scatter(q_accuracy, q_variance)
```
